#### IMPORT REQUIRED LIBRARIES

```python
import pandas as pd
from sklearn import tree
import graphviz
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_val_score
```

#### IMPORT DATASET

```python
df = pd.read_csv("iris.csv")
df
```

#### BUILDING MODEL

```python
x = df[['Sepal_length(cm)', 'Sepal_width(cm)','Petal_length(cm)','Petal_width(cm)']]
y = df['class']
```

```python
#Create an instance of the decision tree classifier
dec_tree = DecisionTreeClassifier(criterion='gini',min_samples_split=2)

#Fit the model with the training data X_train and y_train
dec_tree.fit(x, y)
tree.plot_tree(dec_tree)
```

#### VISUALISATION

```python
data = tree.export_graphviz(dec_tree, out_file= None, filled= True)
graph = graphviz.Source(data)
graph
```
